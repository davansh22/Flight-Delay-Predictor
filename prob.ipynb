```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('flights.csv')
N=15

top_airports = data['ORIGIN_AIRPORT'].value_counts().head(N)


mean_arrival_delays = []
mean_departure_delays = []

for airport in top_airports.index:
    subset = data[data['ORIGIN_AIRPORT'] == airport]
    mean_arrival_delays.append(subset['ARRIVAL_DELAY'].mean())
    mean_departure_delays.append(subset['DEPARTURE_DELAY'].mean())


plt.figure(figsize=(12, 6))  
plt.bar(top_airports.index, mean_arrival_delays, label='Mean Arrival Delay')
plt.bar(top_airports.index, mean_departure_delays, label='Mean Departure Delay', alpha=0.7)
plt.xlabel('Origin Airport')
plt.ylabel('Mean Delay (minutes)')
plt.title(f'Mean Arrival and Departure Delays for Top {N} Origin Airports')
plt.legend()


plt.xticks(rotation=45)

# Show the plot
plt.grid(True)
plt.show()
```

```python
import tensorflow as tf
import pandas as pd

data2 = pd.read_csv('flights.csv')


sample_size = 100000
data_subset = data2.head(sample_size)




X_train_day_of_week = data_subset['DAY_OF_WEEK'].values.reshape(-1, 1)  # Reshape for TensorFlow
y_departure_delay = data_subset['DEPARTURE_DELAY'].values


model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(1,)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(1)
])


model.compile(optimizer='adam', loss='mean_squared_error')


model.fit(X_train_day_of_week, y_departure_delay, epochs=10, batch_size=32) 


input_data = pd.DataFrame({'DAY_OF_WEEK': [1, 2, 3]})  
input_data_numpy = input_data.values.reshape(-1, 1)  

departure_delay_predictions = model.predict(input_data_numpy)
print("Predicted Departure Delays:", departure_delay_predictions)

```

```python
from sklearn.tree import DecisionTreeRegressor
import pandas as pd

data = pd.read_csv('flights.csv')


sample_size = 100000
data_subset = data.head(sample_size)

X_train_features = data_subset[['DAY_OF_WEEK', 'DISTANCE','AIR_TIME']]
y_departure_delay = data_subset['DEPARTURE_DELAY']

y_departure_delay = y_departure_delay.fillna(0)  


tree_model = DecisionTreeRegressor()


tree_model.fit(X_train_features, y_departure_delay)

input_data = pd.DataFrame({'DAY_OF_WEEK': [1, 7, 6], 'DISTANCE': [500, 1000, 1500], 'AIR_TIME': [200, 400, 600]})  # Example input for prediction

departure_delay_predictions = tree_model.predict(input_data)
print("Predicted Arrival Delays:", departure_delay_predictions)
```

```python
from sklearn.metrics import mean_absolute_error

true_values = data_subset['DEPARTURE_DELAY'].fillna(0)  

mae = mean_absolute_error(true_values, tree_model.predict(X_train_features))
print(f"Mean Absolute Error (MAE): {mae}")
```
